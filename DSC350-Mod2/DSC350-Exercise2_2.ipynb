```python
# Author: Andres Melendez
# Date: 05 September 2024

# Define the integer n
n = 3

# Create a list for squares
# The list comprehension iterates over integers from 0 to n (inclusive) and computes the square of each integer.
squares = [i**2 for i in range(n + 1)]

# Create a list for cubes
# The list comprehension iterates over integers from 0 to n (inclusive) and computes the cube of each integer.
cubes = [i**3 for i in range(n + 1)]

# Print the results
# Display the list of squares and cubes.
print("Squares:", squares)
print("Cubes:", cubes)
```

```python
import numpy as np

# Define the integer n
n = 3

# Create an array of integers from 0 to n
# np.arange creates an array with values starting from 0 to n (inclusive).
x = np.arange(n + 1)

# Compute the squares of the array elements
# np.power(x, 2) computes the power of each element in the array x to the exponent 2.
squares = np.power(x, 2)

# Compute the cubes of the array elements
# np.power(x, 3) computes the power of each element in the array x to the exponent 3.
cubes = np.power(x, 3)

# Print the results
# Display the array of squares and cubes.
print("Squares:", squares)
print("Cubes:", cubes)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset
# The Iris dataset is a famous dataset in machine learning and statistics.
# It contains measurements of iris flowers and their species.
iris = load_iris()

# Convert the dataset to a pandas DataFrame for easier manipulation
# The DataFrame is created using the feature data and target names.
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target_names[iris.target]  # Add species column

# Print a description of the dataset
# The description includes basic statistics of the dataset, such as mean, std deviation, min, and max values.
print(f"Dataset Description:{iris_df.describe()}")

# Plot Sepal Length vs Sepal Width
# Sepal length (column 1) will be plotted on the x-axis, and Sepal width (column 2) on the y-axis.
plt.figure(figsize=(8, 6))
plt.scatter(iris_df[iris.feature_names[0]], iris_df[iris.feature_names[1]], c=iris.target, cmap='viridis', edgecolor='k')

# Adding labels and title to the plot
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')
plt.colorbar(label='Species')  # Color bar to show species

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

# Load the California housing dataset
# The California housing dataset contains information about housing in California.
# It includes features such as median income, house age, and others.
california = fetch_california_housing()

# Convert the dataset to a pandas DataFrame for easier manipulation
# Create a DataFrame with the feature data and feature names as column headers.
california_df = pd.DataFrame(data=california.data, columns=california.feature_names)

# Print a description of the dataset
# The description includes basic statistics of the dataset, such as mean, std deviation, min, and max values.
print(f"Dataset Description: \n{california_df.describe()}")

# Plot Median Income (column 0) vs House Age (column 1)
# Column 0 is 'MedInc' and Column 1 is 'HouseAge' in the dataset.

# Create a scatter plot with each point marked with a + sign.
plt.figure(figsize=(8, 6))
plt.scatter(california_df['MedInc'], california_df['HouseAge'], marker='+', color='blue')

# Adding labels and title to the plot
plt.xlabel('Median Income (MedInc)')
plt.ylabel('House Age (HouseAge)')
plt.title('Median Income vs House Age')

# Show the plot
plt.show()

```

```python
import numpy as np

# Create an array of single precision floats
# Using np.float32 specifies that the array should contain 32-bit floating point numbers.
# We initialize the array with values ranging from 0 to 9.
array_single_precision = np.arange(10, dtype=np.float32)

# Print the array
# Display the contents of the array to verify its data type and values.
print(f"Array of Single Precision Floats:\n{array_single_precision}")

# Verify the data type of the array
# Checking the dtype attribute to ensure it is np.float32.
print(f"\nData Type of the Array: {array_single_precision.dtype}")

# Example: Perform a simple operation on the array
# Adding 1.5 to each element in the array to demonstrate floating-point operations.
array_modified = array_single_precision + 1.5

# Print the modified array
print(f"\nModified Array (with 1.5 added to each element):\n{array_modified}")
```

```python
import numpy as np

# Create an array of complex numbers
# We specify complex numbers directly. The 'j' or 'J' denotes the imaginary part in Python.
# Example array contains complex numbers with real and imaginary parts.
array_complex = np.array([1+2j, 3+4j, 5+6j, 7+8j])

# Print the array of complex numbers
# Display the contents of the array to verify its structure and values.
print(f"Array of Complex Numbers:\n{array_complex}")

# Verify the data type of the array
# Checking the dtype attribute to ensure it is np.complex128.
print(f"\nData Type of the Array: {array_complex.dtype}")

# Example: Perform a simple operation on the complex array
# Adding a complex number (2+3j) to each element in the array.
array_modified = array_complex + (2+3j)

# Print the modified array
print(f"\nModified Array (with 2+3j added to each element):\n{array_modified}")
```

```python
# Define an array containing the numbers 0 through 8
# np.arange(9) generates an array with values from 0 to 8 (inclusive).
array = np.arange(9)

# Print the original array
print(f"Original Array:\n{array}")

# Select a part of the array from indexes 3 to 7
# Slicing from index 3 to 7 extracts the elements at indexes 3, 4, 5, and 6.
subarray = array[3:8]

# Print the sliced part of the array
print(f"\nSliced Array (indexes 3 to 7):\n{subarray}")

# Choose elements from index 0 to 7 with an increment of 2
# Slicing with a step of 2 selects every second element from index 0 to 7.
stepped_array = array[0:8:2]

# Print the stepped array
print(f"\nArray with elements from index 0 to 7 with an increment of 2:\n{stepped_array}")

# Reverse the array
# Reversing the array can be done by slicing with a step of -1.
reversed_array = array[::-1]

# Print the reversed array
print(f"\nReversed Array:\n{reversed_array}")
```

```python
import numpy as np

# Create a 3D array with shape (2, 3, 4)
# The array is initialized with values from 0 to 23 and reshaped into a 3D array.
array_3d = np.arange(24).reshape(2, 3, 4)

# Print the original 3D array
print(f"Original 3D Array:\n{array_3d}")

# Ravel
# The ravel() function returns a flattened 1D array from the 3D array.
# It does not create a copy of the data but returns a view when possible.
array_raveled = array_3d.ravel()
print(f"\nRaveled Array:\n{array_raveled}")

# Flatten
# The flatten() function also returns a flattened 1D array, but it creates a copy of the data.
array_flattened = array_3d.flatten()
print(f"\nFlattened Array:\n{array_flattened}")

# Setting the shape with a tuple
# Changing the shape of the array using the reshape() function.
# Here, reshaping to (4, 6) creates a 2D array with 4 rows and 6 columns.
array_reshaped = array_3d.reshape(6, 4)
print(f"\nReshaped Array (6x4):\n{array_reshaped}")

# Transpose
# The transpose() function swaps the dimensions of the array.
# For a 3D array with shape (2, 3, 4), the transposed array will have shape (4, 3, 2).
array_transposed = array_3d.transpose()
print(f"\nTransposed Array:\n{array_transposed}")

# Resize
# The resize() function changes the shape of the array in place and can alter the data if the new shape has more elements.
# Here, resizing the array to (3, 8) will require filling in with repeated values if the new size exceeds the original size.
array_resized = np.resize(array_3d, (3, 8))
print(f"\nResized Array (3x8):\n{array_resized}")
```

```python
import numpy as np

# Create a 2D array with shape (2, 12)
# The array is initialized with values from 0 to 23 and reshaped into a 2D array.
array = np.arange(24).reshape(2, 12)

# Print the original 2D array
print(f"Original 2D Array:\n{array}")

# Number of dimensions
# The ndim attribute returns the number of dimensions of the array.
num_dimensions = array.ndim
print(f"\nNumber of Dimensions: {num_dimensions}")

# Count of elements
# The size attribute returns the total number of elements in the array.
count_elements = array.size
print(f"\nCount of Elements: {count_elements}")

# Count of bytes per element
# The itemsize attribute returns the number of bytes used to store each element.
bytes_per_element = array.itemsize
print(f"\nBytes per Element: {bytes_per_element}")

# Full count of bytes
# The nbytes attribute returns the total number of bytes used by the array.
total_bytes = array.nbytes
print(f"\nTotal Count of Bytes: {total_bytes}")
```

```python
import numpy as np

# Create a NumPy array with complex numbers
# The array contains two complex numbers: 1.j + 1 and 2.j + 3
array_complex = np.array([1.j + 1, 2.j + 3])

# Print the original NumPy array
print(f"Original NumPy Array:\n{array_complex}")

# Convert the NumPy array to a Python list
# The tolist() method converts the NumPy array to a nested Python list.
list_from_array = array_complex.tolist()

# Print the resulting Python list
print(f"\nConverted Python List:\n{list_from_array}")
```
